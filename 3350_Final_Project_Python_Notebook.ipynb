# Import Libraries

```python
#import libraries
import pandas as pd
import numpy as np
import requests
from arcgis.gis import GIS
gis = GIS("home")
```

# Data Extraction

```python
#url List
url = 'https://www.cdph.ca.gov/Programs/CCDPHP/DEODC/CLPPB/Pages/BLLMapsTables2019.aspx#'

r = requests.get(url)
df_list = pd.read_html(r.text)
```

```python
#iterate through items
for x in range(len(df_list)):
    display(df_list[x])
```

```python
#only use the first dataframe
df = df_list[0]
#inspect first and last 5 rows
display(df.head(), df.tail())
```

## Data Cleaning

```python
#strip extra spacing
for x in df.columns:
    df[x] = df[x].str.lstrip()
    
display(df.head(), df.tail())
```

```python
#replaces 'Suppressed' to nan value
#replaces '--' to nan value

for x in df.columns:
    df[x] = df[x].replace('Suppressed', np.nan)
    df[x] = df[x].replace('--', np.nan)
    
display(df.head(), df.tail())
```

```python
#add state for accurate geococodding
df['State'] = 'California'
```

```python
#check columns
df.columns
```

```python
#fills nan values in county column with the front value
df['Local  Health Jurisdiction'] = df['Local  Health Jurisdiction'].replace(np.nan, method= 'ffill')    
```

```python
#display dataframe
display(df.head(), df.tail())
```

```python
#inspect last 10 rows
df.tail(10)
```

```python
#remove last 6 rows
#they are not county rows
df.drop(df.tail(6).index, inplace = True)
```

```python
#check out objects
df.info()
```

```python
#change data types
df['BLL <4.5 % (row)'] = df['BLL <4.5 % (row)'].str.rstrip('%').astype('float') / 100.0
df['BLL ≥4.5 to <9.5 % (row)'] = df['BLL ≥4.5 to <9.5 % (row)'].str.rstrip('%').astype('float') / 100.0
df['BLL ≥9.5 %(row)'] = df['BLL ≥9.5 %(row)'].str.rstrip('%').astype('float') / 100.0

df['Blood Lead Level (BLL) <4.5 n'] = pd.to_numeric(df['Blood Lead Level (BLL) <4.5 n'], errors = 'coerce')
df['BLL≥4.5 to <9.5 n'] = pd.to_numeric(df['BLL≥4.5 to <9.5 n'], errors = 'coerce')
df['BLL≥9.5 n'] = pd.to_numeric(df['BLL≥9.5 n'], errors = 'coerce')
df['Totals'] = pd.to_numeric(df['Totals'], errors = 'coerce')


```

```python
df.info()
```

```python
#Final CheckUp
df.head()
```

```python
#enter file to csv
df.to_csv('/arcgis/home/cdph.csv', index = False)
```
